```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np  # Add this line to import NumPy

# Load the CSV data into a DataFrame
creation_data = pd.read_csv('creation_data_phantom.csv')

# Print data types of DataFrame
print(creation_data.dtypes)

# Print info of DataFrame
print(creation_data.info())

# Print number of missing values
print(creation_data.isna().sum())

# Print description of DataFrame
print(creation_data.describe())

# Replace missing values with mean for 'Earnings_YoY_Growth', 'Players_YoY_Growth', 'Tournaments_YoY_Growth'
mean_earnings_growth = creation_data['Earnings_YoY_Growth'].replace([np.inf, -np.inf], np.nan).mean()
mean_players_growth = creation_data['Players_YoY_Growth'].replace([np.inf, -np.inf], np.nan).mean()
mean_tournaments_growth = creation_data['Tournaments_YoY_Growth'].replace([np.inf, -np.inf], np.nan).mean()

# Replace inf values with NaN in the specified columns
creation_data[['Earnings_YoY_Growth', 'Players_YoY_Growth', 'Tournaments_YoY_Growth']] = creation_data[['Earnings_YoY_Growth', 'Players_YoY_Growth', 'Tournaments_YoY_Growth']].replace([np.inf, -np.inf], np.nan)

# Fill NaN values with the mean
creation_data['Earnings_YoY_Growth'].fillna(mean_earnings_growth, inplace=True)
creation_data['Players_YoY_Growth'].fillna(mean_players_growth, inplace=True)
creation_data['Tournaments_YoY_Growth'].fillna(mean_tournaments_growth, inplace=True)

# Visualize the relationship between 'Earnings_YoY_Growth', 'Players_YoY_Growth', 'Tournaments_YoY_Growth' and other variables if needed

# Save the cleaned DataFrame to a new CSV file
creation_data.to_csv('data_creation_cleaned.csv', index=False)

# Find duplicates
duplicates = creation_data.duplicated()
print("Number of duplicate rows:", duplicates.sum())

# Display duplicate rows
duplicate_rows = creation_data[duplicates]
print("Duplicate Rows:")
print(duplicate_rows)

# Remove duplicates
creation_data_no_duplicates = creation_data.drop_duplicates()

# Verify that duplicates are removed
print("Number of rows after removing duplicates:", len(creation_data_no_duplicates))

# Save the DataFrame without duplicates to a new CSV file
creation_data_no_duplicates.to_csv('data_creation_cleaned_no_duplicates.csv', index=False)
```
